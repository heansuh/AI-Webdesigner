````python
from sentence_transformers import SentenceTransformer

# Initialize embedding model on GPU
model = SentenceTransformer('all-MiniLM-L6-v2', device='cuda')

# Combine relevant fields into a single string
query_text = (
    f"{product_details['ProductTitle']} {product_details['ProductDescription']}"
)

# Generate the embedding
query_embedding = model.encode(query_text)

print("Query Embedding Shape:", query_embedding.shape)
print("Query Embedding:", query_embedding)
````

````python
# Read metadata from .jsonl file
database_path = "meta_Electronics_sampled.jsonl"
products = []

with open(database_path, 'r', encoding='utf-8') as file:
    for line in file:
        product = json.loads(line.strip())
        products.append(product)
````

````python
from qdrant_client import QdrantClient
from qdrant_client.http.models import VectorParams

# Initialize vector database client (Qdrant)
client = QdrantClient(":memory:")  # Use ":memory:" for in-memory DB or provide file path for persistence

# Define vector configuration
vector_params = VectorParams(
    size=384,  # Embedding size
    distance="Cosine"  # Similarity metric
)

# Create a collection in Qdrant
client.create_collection(
    collection_name="products",
    vectors_config={"default": vector_params}  # Use "default" for the default vector field
)

print("Collection created successfully!")
````

````python
from qdrant_client.models import PointStruct
import uuid

for product in products:
    # Extract product metadata
    title = product.get("title", "N/A")
    description = " ".join(product.get("description", []))
    price = product.get("price", 0.0)
    average_rating = product.get("average_rating", 0.0)
    review_numbers = product.get("rating_number", 0)
    store = product.get("store", "N/A")

    # Format text for embedding
    text_to_embed = (
        f"Title: {title}\n"
        f"Description: {description}\n"
        f"Price: {price}\n"
        f"Rating: {average_rating}\n"
        f"Reviews: {review_numbers}\n"
        f"Brand: {store}"
    )

    # Generate embedding
    embedding = model.encode(text_to_embed).tolist()  # Ensure it's a list

    # Generate a unique UUID for the product
    unique_id = str(uuid.uuid4())

    # Create a PointStruct object
    point = PointStruct(
        id=unique_id,  # Use a UUID as the unique identifier
        vector={"default": embedding},  # Specify the vector name as "default"
        payload=product  # Store the full product dictionary as metadata
    )

    # Upload the point to Qdrant
    client.upsert(
        collection_name="products",
        points=[point]  # Pass a list of PointStruct objects
    )

    print(f"Processed product with UUID: {unique_id}")

print("Embeddings and metadata uploaded to Qdrant.")
````

````python
# Step 2: Perform search in Qdrant
results = client.search(
    collection_name="products",  # Name of your collection
    query_vector=("default", query_embedding),  # Specify the named vector
    limit=3  # Number of results to retrieve
)

# Initialize an empty list to store results
products_rag = []

# Step 3: Process and save the results
print("Top 3 results:")
for result in results:
    print(f"Score: {result.score}")
    print(f"Product: {result.payload}")  # Payload contains the original product metadata
    print("---")
    
    # Save the product payload into the list
    products_rag.append(result.payload)
````

````python
import re
import json
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

# Input specifications
specifications = f"This product {first_product_metadata['title']} is designed by brand {first_product_metadata['brand']}. Product description: {first_product_metadata['description']}. Average rating: {first_product_metadata['average_rating']}. Number of ratings: {first_product_metadata['rating_number']}. Price: {first_product_metadata['price']}"

# Load Qwen model with memory optimization
model_id = "Qwen/Qwen2.5-Coder-7B-Instruct"
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype="float16",
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    device_map="cuda"
)
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Define LLM pipeline
llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

def generate_enhanced_metadata(specifications):
    prompt = f"""
    Based on the product specifications '{specifications}', generate:
    1. A very creative company name.
    2. A very long "About Us" section that explains the company's vision and values related to keywords in the product description.
    3. A very creative, random, different product title.
    4. Average rating out of 5.0, at a random but high value.
    5. A random number of ratings.
    6. A random affordable price that suits the product.
    Return the output as a JSON object with the following structure:
    {{
        "company_name": "string",
        "about_us": "string",
        "title": "string",
        "rating": "float",
        "rating_number": "int",
        "price": "float"
    }}
    Return only a JSON object and nothing else.
    """
    # Generate raw response
    response = llm_pipeline(prompt, max_new_tokens=1000, temperature=1.6)[0]["generated_text"]

    # print("Raw Response:")
    # print(response)  # Debugging output

    # Extract JSON using regex
    json_match = re.search(r"```json(.*?)```", response, re.DOTALL)
    if json_match:
        json_content = json_match.group(1).strip()  # Get the JSON between ```json and ```
        try:
            # Parse the extracted JSON
            return json.loads(json_content)
        except json.JSONDecodeError as e:
            print(f"Failed to parse JSON: {e}")
            print("Extracted JSON Content:", json_content)  # Debugging malformed JSON
            return None
    else:
        print("No valid JSON object found in the response.")
        return None

# Test the function
enhanced_metadata = generate_enhanced_metadata(specifications)
print(json.dumps(enhanced_metadata, indent=4))
````

````python
def generate_html(metadata):
    prompt = f"""
    Create a simple HTML template for a company website based on the following metadata:
    Company Name: {metadata['company_name']}
    About Us: {metadata['about_us']}
    Product Title: {metadata['title']}
    Product Price: {metadata['price']}
    Product Rating: {metadata['rating']}
    Product Rating Number: {metadata['rating_number']}
    
    The HTML should include:
    - A header with the company name and a placeholder image for the company logo (logo-placeholder.png).
    - An "About Us" section with the provided content.
    - A "Products" section with a table containing product title, price, rating, and number of ratings, along with a placeholder image for the product (product-placeholder.png).
    - A "Contact" section with placeholder contact details.

    Make sure to replace the HTML content below with metadata.
    
    Example HTML:
    <!DOCTYPE html>
    <html>
    <head>
        <title>{metadata['company_name']}</title>
    </head>
    <body>
        <header>
            <h1>{metadata['company_name']}</h1>
            <img src="logo-placeholder.png" alt="Company Logo">
        </header>
        <section>
            <h2>About Us</h2>
            <p>{metadata['about_us']}</p>
        </section>
        <section>
            <h2>Products</h2>
            <table>
                <tr>
                    <th>Title</th>
                    <th>Price</th>
                    <th>Rating</th>
                    <th>Number of Ratings</th>
                    <th>Image</th>
                </tr>
                <tr>
                    <td>{metadata['title']}</td>
                    <td>${metadata['price']}</td>
                    <td>{metadata['rating']}</td>
                    <td>{metadata['rating_number']}</td>
                    <td><img src="product-placeholder.png" alt="Product Image"></td>
                </tr>
            </table>
        </section>
        <section>
            <h2>Contact</h2>
            <p>Email: contact@company.com</p>
            <p>Phone: (123) 456-7890</p>
        </section>
    </body>
    </html>
    Return only valid HTML content.
    """
    response = llm_pipeline(prompt, max_new_tokens=500, temperature=0.8)[0]["generated_text"]

    # Extract valid HTML
    html_match = re.search(r"<!DOCTYPE html>.*</html>", response, re.DOTALL)
    if html_match:
        return html_match.group(0).strip()
    else:
        print("No valid HTML found in the response.")
        print("Raw Response:", response)  # Debugging output
        return None

# Generate and Save HTML
html_output = generate_html(enhanced_metadata)
if html_output:
    output_file = "company_page.html"
    with open(output_file, "w") as file:
        file.write(html_output)
    print(f"Website generated successfully! Open {output_file} in your browser.")
else:
    print("Failed to generate HTML.")
````

````python
### Image Generator

import torch

from diffusers import UniDiffuserPipeline
from diffusers.utils import load_image

device = "cuda"
model_id_or_path = "thu-ml/unidiffuser-v1"
pipe = UniDiffuserPipeline.from_pretrained(model_id_or_path, torch_dtype=torch.float16)
pipe.to(device)

# Image variation can be performed with an image-to-text generation followed by a text-to-image generation:
# 1. Image-to-text generation
image_url = first_product_metadata['images']
init_image = load_image(image_url).resize((512, 512))

sample = pipe(image=init_image, num_inference_steps=30, guidance_scale=8.0)
i2t_text = sample.text[0]
print(i2t_text)

if first_product_metadata['brand'] in i2t_text:
    # Replace the brand with the enhanced company name
    updated_text = i2t_text.replace(first_product_metadata['brand'], enhanced_metadata['company_name'])
    print("Brand replaced in text:", updated_text)
else:
    # Concatenate enhanced company name with i2t_text
    updated_text = enhanced_metadata['company_name'] + " " + i2t_text
    print("Brand not found, concatenating company name:", updated_text)

# 2. Text-to-image generation
sample = pipe(prompt=updated_text, num_inference_steps=50, guidance_scale=7.0)
final_image = sample.images[0].resize((128,128))
final_image.save("product-placeholder.png")
````

````python
from diffusers import StableDiffusionXLPipeline
from diffusers import EulerAncestralDiscreteScheduler
import torch

pipe = StableDiffusionXLPipeline.from_single_file("https://huggingface.co/martyn/sdxl-turbo-mario-merge-top-rated/blob/main/topRatedTurboxlLCM_v10.safetensors")
pipe.to("cuda")
pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)

# Load the LoRA
pipe.load_lora_weights('ntc-ai/SDXL-LoRA-slider.futuristic-logo-design', weight_name='futuristic logo design.safetensors', adapter_name="futuristic logo design")

# Activate the LoRA
pipe.set_adapters(["futuristic logo design"], adapter_weights=[2.0])

prompt = f"{enhanced_metadata['company_name']}, electronics, futuristic logo design"
negative_prompt = "nsfw"
width = 512
height = 512
num_inference_steps = 50
guidance_scale = 7.5
image = pipe(prompt, negative_prompt=negative_prompt, width=width, height=height, guidance_scale=guidance_scale, num_inference_steps=num_inference_steps).images[0]
image.save('logo-placeholder.png')
````
